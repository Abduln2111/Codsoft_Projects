#Data Visulaisation 

```python
# Calculate the correlation matrix for all features
correlation_matrix = hrdata.corr()

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Plot the count of each MaritalStatus category
plt.figure(figsize=(10, 6))
sns.countplot(data=hrdata, x='MaritalStatus', palette='viridis')
plt.title('Distribution of Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

```python
# Plot the count of each JobRole category
plt.figure(figsize=(12, 6))
sns.countplot(data=hrdata, x='JobRole', palette='viridis')
plt.title('Distribution of Job Roles')
plt.xlabel('Job Role')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python

# Check if the dataset has 'YearsAtCompany' and 'Gender' columns
if 'YearsAtCompany' in hrdata.columns and 'Gender' in hrdata.columns:
    # Create a DataFrame with the count of each gender over 'YearsAtCompany'
    gender_years_count = hrdata.groupby(['YearsAtCompany', 'Gender']).size().unstack().fillna(0)

    # Plot the line graph
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=gender_years_count)
    plt.title('Gender Distribution Over Years at Company')
    plt.xlabel('Years at Company')
    plt.ylabel('Count')
    plt.legend(title='Gender')
    plt.show()
else:
    print("The dataset does not contain 'YearsAtCompany' or 'Gender' columns. Please provide relevant columns.")
```

```python
# Plot the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=hrdata, x='TotalWorkingYears', y='EducationField', alpha=0.6, s=100, palette='viridis')
plt.title('Total Working Years vs EducationField')
plt.xlabel('Total Working Years')
plt.ylabel('EducationField')
plt.grid(True)
plt.show()
```

```python
# Calculate value counts for OverTime
overtime_counts = hrdata['OverTime'].value_counts()

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(overtime_counts, labels=overtime_counts.index, autopct='%1.1f%%', startangle=140, colors=['lightsalmon', 'skyblue'])
plt.title('Distribution of OverTime')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Plotting the line chart
plt.figure(figsize=(10, 6))
sns.lineplot(data=hrdata, x='NumCompaniesWorked', y='DistanceFromHome', marker='o', color='skyblue')
plt.title('Average DistanceFromHome vs NumCompaniesWorked')
plt.xlabel('NumCompaniesWorked')
plt.ylabel('Average DistanceFromHome')
plt.grid(True)
plt.show()
```
